```python
import cv2

# Select tracking algorithm
tracker = cv2.TrackerCSRT_create()  # Other options: TrackerKCF_create(), TrackerMIL_create()

# Load video file or use webcam (0 for default webcam)
video_path = r"C:\Users\FINRISE\Desktop\object tracking\video\Traffic.mp4"  # Change to your video path
cap = cv2.VideoCapture(video_path)

# Check if video loaded
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Read the first frame
ret, frame = cap.read()
if not ret:
    print("Error: Could not read the first frame.")
    exit()

# Select a bounding box for tracking
bbox = cv2.selectROI("Select Object", frame, False)
tracker.init(frame, bbox)

# Start tracking loop
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Stop if video ends

    # Update tracker
    success, bbox = tracker.update(frame)

    # Draw bounding box
    if success:
        x, y, w, h = map(int, bbox)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)
        cv2.putText(frame, "Tracking", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    else:
        cv2.putText(frame, "Lost Tracking", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    # Show result
    cv2.imshow("Object Tracking", frame)

    # Exit on 'q' key press
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2

# Check OpenCV version and use the correct tracker
if int(cv2.__version__.split(".")[0]) < 4:  # If OpenCV version is below 4.x
    tracker = cv2.TrackerCSRT_create()
else:
    tracker = cv2.legacy.TrackerCSRT_create()  # Use legacy module for OpenCV 4.x

# Load video file or use webcam (0 for default webcam)
video_path = r"C:\Users\FINRISE\Desktop\object tracking\video\Traffic.mp4"  # Change to your video path
output_path = r"C:\Users\FINRISE\Desktop\object tracking\output\tracked_output.mp4"  # Output path

cap = cv2.VideoCapture(video_path)

# Check if video loaded
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Get video properties
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Define codec and create VideoWriter object
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec for .mp4 format
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))

# Read the first frame
ret, frame = cap.read()
if not ret:
    print("Error: Could not read the first frame.")
    exit()

# Select a bounding box for tracking
bbox = cv2.selectROI("Select Object", frame, False)
tracker.init(frame, bbox)

# Start tracking loop
while True:
    ret, frame = cap.read()
    if not ret:
        break  # Stop if video ends

    # Update tracker
    success, bbox = tracker.update(frame)

    # Draw bounding box
    if success:
        x, y, w, h = map(int, bbox)
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 255, 0), 3)
        cv2.putText(frame, "Tracking", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
    else:
        cv2.putText(frame, "Lost Tracking", (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    # Save the frame to the output video file
    out.write(frame)

    # Show result
    cv2.imshow("Object Tracking", frame)

    # Exit on 'q' key press
    if cv2.waitKey(30) & 0xFF == ord('q'):
        break

# Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Tracking output saved at: {output_path}")
```
